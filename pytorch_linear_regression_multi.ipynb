```python
    import torch
    import torch.autograd
    import torch.nn.functional as F
    from torch.autograd import Variable
    import torch.nn as nn
    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn.preprocessing import PolynomialFeatures
```

```python
train_data = torch.Tensor([
   [40,  6,  4],
   [44, 10,  4],
   [46, 12,  5],
   [48, 14,  7],
   [52, 16,  9],
   [58, 18, 12],
   [60, 22, 14],
   [68, 24, 20],
   [74, 26, 21],
   [80, 32, 24]])
test_data = torch.Tensor([
    [6, 4],
    [10, 5],
    [4, 8]])

x_train = train_data[:,1:3]
y_train = train_data[:,0]
num_epochs = 1000
POLY_DEGREE = 3
input_size = 2
output_size = 1
```

```python
poly = PolynomialFeatures(input_size * POLY_DEGREE, include_bias=False)
x_train_poly = poly.fit_transform(x_train.numpy())
```

```python

class Model(torch.nn.Module):

    def __init__(self):
        super(Model, self).__init__()
        #self.linear1 = torch.nn.Linear(poly.n_output_features_, output_size)
        self.linear1 = torch.nn.Linear(poly.n_output_features_, output_size)
        
                
    def forward(self, x):
        out = self.linear1(x)
        return out
            
model = Model()    
criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)

losses = [] # Added
```

```python
for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = model(Variable(torch.Tensor(x_train_poly)))
    #outputs = model(Variable(x_train))
    loss = criterion(outputs,Variable(y_train))
    losses.append(loss.data[0])
    loss.backward()
    optimizer.step()
    
    if loss.data[0] < 1e-4:
        break
    if(epoch+1) %5 ==0:
        print('epoch [%d/%d], Loss: %.4f' % (epoch +1, num_epochs, loss.data[0]))
        plt.plot(losses)
        plt.show()
        #predicted = model(Variable(torch.from_numpy(x_train))).data.numpy()
        #plt.plot(x_train,y_train,'ro',label='Original Data')
        #plt.plot(x_train,predicted,label='Fitted Line')
        #plt.legend()
        #plt.show()
```
